### Imports

```python
import torch
import numpy as np

from code.manifold_vae.lib.trainer import EuclideanVAETrainer
from code.manifold_vae.lib.dataloaders import load_fashion_mnist
from code.manifold_vae.lib.utils import plot_losses
```


### Set up and initialize data loader

```python
# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)
```

```python
latent_dim = 20
num_epochs = 2
learning_rate = 0.001
device = "cpu"

vae_config = { 'latent_dim': latent_dim, 'num_epochs': num_epochs, 'learning_rate': learning_rate, 'device': device }


batch_size = 200
dataloader = load_fashion_mnist(batch_size)
```

### Train and evaluate model

```python
train_losses, test_losses = EuclideanVAETrainer(vae_config,dataloader).train()
```

### Visualization 

```python
plot_losses(train_losses, test_losses, num_epochs)
```
