### Imports

```python
import sys
import os

sys.path.append(os.path.abspath('..'))
print(sys.path)

import torch
import numpy as np

from .trainer import EuclideanVAETrainer
from .dataloaders import load_fashion_mnist
from .utils import plot_losses
```


### Set up and initialize data loader

```python
# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)
```

```python
latent_dim = 20
num_epochs = 10
learning_rate = 0.001
device = "cpu"

vae_config = { 'latent_dim': latent_dim, num_epochs: num_epochs, 'learning_rate': learning_rate, 'device': device }


batch_size = 10
dataloader = load_fashion_mnist(batch_size)
```

### Train and evaluate model

```python
train_losses, test_losses = EuclideanVAETrainer(vae_config,dataloader).train()
```

### Visualization 

```python
plot_losses(train_losses, test_losses)
```
